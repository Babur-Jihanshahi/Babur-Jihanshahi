```python
# Quadratic equation:

def quad(a, b, c):
  if b**2 - 4*a*c > 0:
    x1 = (-b - math.sqrt(b**2 - 4*a*c)) / 2*a
    x2 = (-b + math.sqrt(b**2 - 4*a*c)) / 2*a
    print(f"The answers are {x1} and {x2}")

  elif b**2 - 4*a*c == 0:
    x = -b / 2*a
    print(f"There are two same root {x}")

equations = [[1, 1000.001, 1], [1, -4, 3.999999], [1.002, 11.01, 0.01265]]
for e in equations:
  quad(e[0], e[1], e[2])
```

```python
# Solving binomial coefficient

def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

def binomial(a, b):
  return factorial(a) / (factorial(b) * factorial(a - b))

binomial_coefficients = [[50, 46], [100, 98]]
for e in binomial_coefficients:
  print(binomial(e[0], e[1]))
```

```python
# Trapez rule

def trap(f, a, b, n):

  x = np.linspace(a, b, n + 1)
  y = f(x)
  h = (b - a) / n


  integral = h * (0.5 * y[0] + 0.5 * y[-1] + np.sum(y[1: -1]))
  return integral

def simp(f, a, b, n):
  x = np.linspace(a, b, n + 1)
  y = f(x)
  h = (b - a)/ n

  integral = h/3 * (y[0] + y[-1] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-1:2]))
  return integral

f = lambda x: 3*x**3 + 5*x -1
a = 0
b = 1
n = [4, 8, 16]
for values in n:
  integral_value_1 = trap(f, a, b, values)
  integral_value_2 = simp(f, a, b, values)
  print(integral_value_1, integral_value_2)
```

# Debye Model

The Debye model describes the phonon contribution to the specific heat of a solid. It is an improvement over the Einstein model, especially at low temperatures.

## Specific Heat in the Debye Model

The formula for the specific heat $(C_v$) in the Debye model is:

$$
C_v = 9k_B \left(\frac{T}{\Theta_D}\right)^3 \int_0^{\Theta_D/T} \frac{x^4 e^x}{(e^x-1)^2} dx
$$

Where:

- $( k_B $) is the Boltzmann constant.
- $( T $) is the temperature.
- $( \Theta_D $) is the Debye temperature, a characteristic temperature of the material related to the Debye frequency.

At low temperatures, the specific heat predicted by the Debye model varies as $(T^3$), which aligns well with experimental observations.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
k_b = 1.38e-23  # Boltzmann constant
theta_d = 300   # Debye temperature

T = np.arange(0, 301, 5, dtype=float)  # Starting from 1 to avoid division by zero
N = 100
c_v = []

def f(x):
    if x < 1e-10:  # Handling the x=0 case
        return x**2 * (1 + x)
    else:
        return x**4 * np.exp(x) / (np.exp(x) - 1)**2

def trapez(a, b, n):
    if n < 1 or b < a:
        print("Wrong Input")
    else:
        h = (b - a) / n
        s = 0.5 * (f(a) + f(b))
        for i in range(1, n):
            s += f(a + i * h)

    return h * s

for temps in T:
    constants = 9 * k_b * (temps / theta_d)**3
    c_v_value = constants * trapez(0, theta_d / temps, N)
    c_v.append(c_v_value)

    # Print the computed c_v values
    print("T =", temps, ", c_v =", c_v_value)

c_v_arr = np.array(c_v)

# Plotting
plt.plot(T, c_v_arr)
plt.xlabel('Temperature (K)')
plt.ylabel('Specific Heat c_v (J/K)')
plt.title('Debye Model')
plt.grid(True)
plt.show()

```

# Area of an elipse
- Equation for the elipse is given as $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$
- The are of the elipse is given as:
$$S = 4 \int_0^a y(x)dx$$

```python
import numpy as np

def f(x, a, b):
  return b * np.sqrt(1 - x**2 / a**2)

def trapez(func, a, b, n, *args):
  if n < 1 or b < a:
    print('Wrong Input')
  else:
    h = (b - a) / n
    s = 0.5 * (f(a, *args) + f(b, *args))
    for i in range(1, n):
      s += f(a + i * h, *args)

  return h*s

def elips_area(a, b, n):
  return 4 * trapez(f, 0, a, n, a, b)

def real_elips(a, b):
  return np.pi * a * b

a = 2
b = 1
n = 1000
num_elipse = elips_area(a, b, n)
ana_elipse = real_elips(a, b)

print(f'Error between the numerical and analytical calculation: {ana_elipse - num_elipse}')


```

# Fresnel Integral
- We know that when light passes through a rectangular obstacle it forms diffraction fringes. The expression for the intensity of light at a distance d from one edge of the obstacle is
$$I = \frac{I_0}{2} \{[C(d) + 0.5]^2 + [S(d) + 0.5]^2\}$$
- Here $I_0$ is the luminous intensity of the source and $C(d)$ and $S(d)$ are Fresnel integrals.
$$C(d) = \int_0^d cos(\frac{\pi}{2} x^2)$$
$$S(d) = \int_0^d sin(\frac{\pi}{2} x^2)$$
- Write a program that calculates the Fresnel integrals for a given distance $d$ using Simpson's formula and graph the variation of the relative luminous intensity $(I/I_0)$ over the interval $d = -1, +4$ in steps $0.1$

```python
import numpy as np
import matplotlib.pyplot as plt

# define the equations
def c(x):
  return np.cos((np.pi/2) * (x**2))

def s(x):
  return np.sin((np.pi/2) * (x**2))

def simpson(func, a, b, n):
  if n%2 != 0:
    print('N should be even!')
    return None

  h = (b - a) / n
  s = func(a) + func(b)
  for i in range(1, n, 2):
    s += 4 * func(a + i * h)
  for i in range(2, n - 1, 2):
    s += 2 * func(a + i * h)

  return h * s / 3

d_values = np.arange(-1, 4.1, 0.001)
light_intensity = []
n = 1000

for d in d_values:
  if d >= 0:
    C_d = simpson(c, 0, d, n)
    S_d = simpson(s, 0, d, n)
  else:
    C_d = -simpson(c, d, 0, n)
    S_d = -simpson(s, d, 0, n)

  l_intense = 0.5 * ((C_d + 0.5)**2 + (S_d + 0.5)**2)
  light_intensity.append(l_intense)

light_intensity_arr = np.array(light_intensity)

# Plotting
plt.plot(d_values, light_intensity_arr)
plt.xlabel('Distance d')
plt.ylabel('I/I_0')
plt.title('Fresnel Integrals')
plt.grid(True)
plt.show()
```
